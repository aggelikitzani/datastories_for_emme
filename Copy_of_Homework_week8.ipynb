**Άσκηση 9**:

Βρείτε το συνολικό πληθυσμό (sum) σε κάθε ήπειρο

```python
# γκρουπάρισμα ανά ήπειρο και εύρεση του συνολικού πληθυσμού της
import pandas as pd

data = {
    'country': ['Afghanistan', 'Bangladesh', 'Bhutan', 'India', 'Japan', 'Nepal', 'Pakistan', 'Philippines', 'Sri Lanka', 'Vietnam'],
    'continent': ['Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia'],
    'population': [38928346, 164689383, 771608, 1380004385, 126476461, 29136808, 220892331, 109581078, 21413249, 97338579]
}

# Δημιουργία DataFrame
df = pd.DataFrame(data)

# Ομαδοποίηση ανά ήπειρο και εύρεση του συνολικού πληθυσμού
total_population_per_continent = df.groupby('continent')['population'].sum()

print(total_population_per_continent)
```

**Άσκηση 10**:

Βρείτε σε ποιες ηπείρους ο μέσος όρος του προσδόκιμου ζωής είναι μικρότερος από 55

```python
import pandas as pd

data = {
    'country': ['Afghanistan', 'Bangladesh', 'Bhutan', 'India', 'Japan', 'Nepal', 'Pakistan', 'Philippines', 'Sri Lanka', 'Vietnam'],
    'continent': ['Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia'],
    'life_expectancy': [59.3, 72.3, 70.6, 69.4, 84.5, 70.2, 66.5, 71.0, 76.9, 75.3]
}

# Δημιουργία DataFrame
df = pd.DataFrame(data)

# Ομαδοποίηση ανά ήπειρο και υπολογισμός του μέσου όρου του προσδόκιμου ζωής
mean_life_expectancy_per_continent = df.groupby('continent')['life_expectancy'].mean()

# Φιλτράρισμα για μέσο όρο προσδόκιμου ζωής μικρότερο από 55
continents_with_low_life_expectancy = mean_life_expectancy_per_continent[mean_life_expectancy_per_continent < 55]

print(continents_with_low_life_expectancy)
```

**Άσκηση 11**:

Βρείτε πόσες χώρες υπάρχουν σε κάθε ήπειρο (count) και ταξινομήστε το αποτέλεσμα σε φθίνουσα σειρά


```python
import pandas as pd

data = {
    'country': ['Afghanistan', 'Bangladesh', 'Bhutan', 'India', 'Japan', 'Nepal', 'Pakistan', 'Philippines', 'Sri Lanka', 'Vietnam'],
    'continent': ['Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia', 'Asia'],
}

# Δημιουργία DataFrame
df = pd.DataFrame(data)

# Ομαδοποίηση ανά ήπειρο και καταμέτρηση των χωρών
country_count_per_continent = df.groupby('continent')['country'].count()

# Ταξινόμηση των αποτελεσμάτων σε φθίνουσα σειρά
sorted_country_count_per_continent = country_count_per_continent.sort_values(ascending=False)

print(sorted_country_count_per_continent)
```
